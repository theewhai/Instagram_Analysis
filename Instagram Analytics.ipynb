```python
import pandas as pd

df = pd.read_csv("C:/Users/HP/Downloads/Instagram_Analytics.csv")

df.head(10)
```

```python
df["upload_date"] = pd.to_datetime(df["upload_date"]).dt.floor('S')
```

```python
df["total_engagement"] = (
    df["likes"] +
    df["comments"] +
    df["shares"] +
    df["saves"]
)
```

```python
#Engagement Metric
df.head()
```

```python
#Performance Ratio

df["virality_rate"] = df["shares"] / df["impressions"]
df["saves_rate"] = df["saves"] / df["impressions"]
df["reach_rate"] = df["reach"] / df["impressions"]
```

```python
df.head()
```

```python
df["content_category"].dtype
```

```python
corr_cols = [
    "likes", "comments", "shares", "saves", "reach", "impressions", "followers_gained", "engagement_rate"
]
correlation_matrix = df[corr_cols].corr()
correlation_matrix
```

```python
df.head()
```

```python
df["post_day"] = df["upload_date"].dt.day_name()
df["post_month"] = df["upload_date"].dt.to_period("M").astype(str)
df["post_hour"] = df["upload_date"].dt.hour
```

```python
df.head()
```

```python
media_perf = (
    df.groupby("media_type")
        .agg(
            avg_engagement_rate = ("engagement_rate", "mean"),
            avg_reach = ("reach", "mean"),
            avg_followers_gained = ("followers_gained", "mean")
        )
        .reset_index()
        .sort_values("avg_engagement_rate", ascending = False)
)

media_perf
```

```python
#Best posting days

day_perf = (
    df.groupby("post_day")
    .agg(avg_engagement_rate = ("engagement_rate", "mean"))
    .reset_index()
    .sort_values("avg_engagement_rate", ascending = False)
)

day_perf
```

```python
category_perf = (
    df.groupby("content_category")
        .agg(
            avg_engagement_rate = ("engagement_rate", "mean"),
            avg_followers_gained = ("followers_gained", "mean")
        )
    .reset_index()
    .sort_values("avg_engagement_rate", ascending = False)
)

category_perf
```
